```python
import pyxdf
```

```python
# Specify the path to the xdf file
file_path = '/home/ubuntu/portiloop-training/temp/sub-Hugo_Test_ses-S001_task-Default_run-001_eeg.xdf'

# Open the xdf file
streams, header = pyxdf.load_xdf(file_path)

# Print the streams and header
print("Streams:")
print(streams)
print("\nHeader:")
print(header)
```

```python
for stream in streams:
    stream_name = stream['info']['name'][0]
    num_channels = stream['info']['channel_count'][0]
    print(f"Stream name: {stream_name}")
    print(f"Number of channels: {num_channels}")
```

```python
portiloop_stream = [stream for stream in streams if stream['info']['name'][0] == 'Portiloop Filtered'][0]
actichamp_stream = [stream for stream in streams if stream['info']['name'][0] == 'actiCHamp-22110882'][0]
portiloop_raw_stream = [stream for stream in streams if stream['info']['name'][0] == 'Portiloop Raw Data'][0]
```

```python
import numpy as np
import os

save_path = '/home/ubuntu/portiloop-training/temp/'

# Save the time series from portiloop_stream
portiloop_data = portiloop_stream['time_series']
np.savez(os.path.join(save_path, 'portiloop_data.npz'), data=portiloop_data)

# Save the time series from actichamp_stream
actichamp_data = actichamp_stream['time_series']
np.savez(os.path.join(save_path, 'actichamp_data.npz'), data=actichamp_data)

# Save the time series from portiloop_raw_stream
portiloop_raw_data = portiloop_raw_stream['time_series']
np.savez(os.path.join(save_path, 'portiloop_raw_data.npz'), data=portiloop_raw_data)
```
